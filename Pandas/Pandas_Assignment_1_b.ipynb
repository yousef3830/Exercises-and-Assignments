#### #7. Filter the data set. Create three sub DataFrames from the `position` column for `G`, `F`, and `C`.

```python
guards = nba_df[nba_df['pos'] == 'G']
forwards = nba_df[nba_df['pos'] == 'F']
centers = nba_df[nba_df['pos'] == 'C']

print("Guards:")
print(guards.head())

print("Forwards:")
print(forwards.head())

print("Centers:")
print(centers.head())
```

#### #8.  Run `describe()` on these new DataFrames. Compare the mean field goals (the `fg` column) between positions.

```python
guards_stats = guards.describe()
print("Mean field goals for Guards:", guards_stats.loc['mean', 'fg'])
```

```python
forwards_stats = forwards.describe()
print("Mean field goals for Forwards:", forwards_stats.loc['mean', 'fg'])
```

```python
centers_stats = centers.describe()
print("Mean field goals for Centers:", centers_stats.loc['mean', 'fg'])
```
